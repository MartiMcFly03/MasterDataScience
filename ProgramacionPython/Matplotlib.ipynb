# Trabajando con matplotlib en jupyter

```python
#Trabajando con matplotlib dento de jupyter

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#grafica lineal sencilla
values = np.array([1,2,3,4,5])
plt.plot(values)
plt.show()
```

Y si le pasamos 2 conjuntos diferentes, toma uno para x y otro para y

```python
#Pasandole 2 conjuntos de datos

values_x = np.array([1,2,3,4,5])
values_y = np.array([1,4,9,16,25])
plt.plot(values_x, values_y)
plt.show()
```

podemos cambiar la forma de los puntos, por ejemplo a pentagonos rojos usando 'pr' como parametro 

```python
#usando pentagonos rojos para los puntos.

values_x = np.array([1,2,3,4,5])
values_y = np.array([1,4,9,16,25])
plt.plot(values_x, values_y, 'rp')
plt.show()
```

Usando linea descontinua y aumentando el grosor

```python
values_x = np.array([1,2,3,4,5])
values_y = np.array([1,4,9,16,25])
plt.plot(values_x, values_y, 'r--', linewidth=4)
plt.show()
```

## Imprimiendo multiples graficas 

```python
#Imprimiendo multiples funciones en una imagen
#graficando una funcion cuadratica
x = np.arange(-10, 11)
y = x**2
plt.plot(x, y, '--')

#graficando una funcion lineal
y2 = 2*x + 5
plt.plot(x, y2, 'g-')

plt.show()
```

## Generando un Histograma


```python
#Generando un histograma de 1000 valores aleatorios con 100 bins (divisiones)

values = np.random.randn(1000) # 1000 valores aleatorios
plt.hist(values, bins=20)
plt.show()
#print(values)
```

### Histogramas con Parametros extra

```python
#Haciendo un histograma rojo de forma horizontal con 20 bins y 1000 valores aleatorios
values = np.random.randn(1000) # 1000 valores aleatorios
plt.hist(values, bins=20, orientation='horizontal', color='r') # histograma horizontal en rojo con 20 bins
plt.show()
```

```python
#Modificando el rango de valores del histograma
values = np.random.randn(1000) # 1000 valores aleatorios
plt.hist(values, bins=20, range=(-3,3)) # histograma con rango de -3 a 3 y 20 bins
plt.show()
```

### Histograma dos dimensiones

```python
#Generando un histograma de 2 dimensiones con 1000 valores aleatorios en x y 1000 valores aleatorios en y 
x = np.random.randn(1000) # 1000 valores aleatorios en x
y = np.random.randn(1000) # 1000 valores aleatorios en y
plt.hist2d(x, y, bins=30) # histograma 2D con 30x30 bins
plt.show()

```

## Grafica de barras

```python
#generando una grafica de barras con 5 barras de diferentes alturas y anchos de 0.5 

fig = plt.figure() # se crea una figura
ax = fig.add_axes([0,0,1,1]) # se agrega un eje a la figura con coordenadas de 0 a 1 en x y en y

values_x = ['a', 'b', 'c', 'd', 'e']
values_y = np.random.randint(low=1, high=10, size=5) # 5 valores aleatorios entre 1 y 10

ax.bar(values_x, values_y, width=0.5) # grafica de barras con 5 barras de diferentes alturas y anchos de 0.5
plt.show()
```

### Multiples graficos de barras en 1 imagen

```python
#Agregando multiples graficas de barras a una misma imagen

#data to plot
n_groups = 10 # 5 grupos de barras en la grafica
values_1 = np.random.randint(low=1, high=100, size=10) # 5 valores aleatorios entre 1 y 10
values_2 = np.random.randint(low=1, high=100, size=10) # 5 valores aleatorios entre 1 y 10

# create plot
fig, ax = plt.subplots() # se crea una figura y un eje para la grafica
index = np.arange(n_groups) # se crea un arreglo de 5 elementos
bar_width = 0.35 # ancho de las barras

# se grafican las barras
rects1 = plt.bar(index, values_1, bar_width, label='Values 1', color='b') # barras azules con los valores 1 
#rects2 = plt.bar(index + bar_width, values_2, bar_width, label='Values 2', color='r') # barras rojas con los valores 2
rects2 = plt.bar(index, values_2, bar_width, label='Values 2', color='r') # barras rojas con los valores 2

# se agregan etiquetas a la grafica
plt.xlabel('Group') # etiqueta en x
plt.ylabel('Values') # etiqueta en y
plt.title('Values by group') # titulo de la grafica
#plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E')) # etiquetas en x
plt.xticks(index, ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J')) # etiquetas en x
plt.legend() # se agrega la leyenda a la grafica

# se muestra la grafica
plt.tight_layout() # se ajusta la grafica para que se muestre correctamente en la imagen final
plt.show() # se muestra la grafica
```

Como vemos en el ejemplo anterior al graficar "rects2" el index es el mismo que el index de "rects1" razon por la cual las graficas quedan encimadas una tras la otra.
esto se corrije sumandole al index de la grafica "rects2" el ancho de la grafica "rects1"

Vamos a corregirlo.

```python
#Agregando multiples graficas de barras a una misma imagen

#data to plot
n_groups = 10 # 5 grupos de barras en la grafica
values_1 = np.random.randint(low=1, high=100, size=10) # 5 valores aleatorios entre 1 y 10
values_2 = np.random.randint(low=1, high=100, size=10) # 5 valores aleatorios entre 1 y 10

# create plot
fig, ax = plt.subplots() # se crea una figura y un eje para la grafica
index = np.arange(n_groups) # se crea un arreglo de 5 elementos
bar_width = 0.35 # ancho de las barras

# se grafican las barras
rects1 = plt.bar(index, values_1, bar_width, label='Values 1', color='b') # barras azules con los valores 1 
rects2 = plt.bar(index + bar_width, values_2, bar_width, label='Values 2', color='r') # barras rojas con los valores 2

# se agregan etiquetas a la grafica
plt.xlabel('Group') # etiqueta en x
plt.ylabel('Values') # etiqueta en y
plt.title('Values by group') # titulo de la grafica
#plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E')) # etiquetas en x
plt.xticks(index, ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J')) # etiquetas en x
plt.legend() # se agrega la leyenda a la grafica

# se muestra la grafica
plt.tight_layout() # se ajusta la grafica para que se muestre correctamente en la imagen final
plt.show() # se muestra la grafica
```

## Trabajando con Graficas de pastel

```python
# Grafico de pastel sencillisimisimo

values = np.array([0.2, 0.3, 0.4, 0.1]) # valores para el grafico de pastel (20%, 30%, 40%, 10%) suman 100%
plt.pie(values) # grafico de pastel
plt.show() # se muestra el grafico
```

Si los valores sumados no cumplen el 100% se hara la interpolacion automaticamente
donde la suma de los valores sera el 100%
y una regla de 3 determinará el porcentaje de cada seccion.

```python
# Grafico de pastel sencillisimisimo

values = np.array([0.2, 0.1, 0.2, 0.1]) # valores para el grafico de pastel (20%, 10%, 20%, 10%) suman 60% que se transforma a 100% en el grafico
plt.pie(values) # grafico de pastel
plt.show() # se muestra el grafico
```

Podemos agregar etiquetas a cada seccion del grafico y modificar el radio.

```python
# Generando un grafico de pastel con 5 secciones

values = np.random.randint(low=1, high=10, size=5) # 5 valores aleatorios entre 1 y 10
print(values)
labels = ['A', 'B', 'C', 'D', 'E'] # etiquetas para las secciones
plt.pie(values, labels=labels, autopct='%1.1f%%', radius=2) # grafico de pastel con los valores y etiquetas dadas y porcentajes de 1 decimal y radio de 2
plt.show() # se muestra la grafica
```

## Multiples Graficos
Se pueden agregar multiples graficos en una sola imagen usando la funcion subplot()

```python
#Generando 3 graficas diferentes en una misma imagen con subplots
%matplotlib inline
from ast import Import
import matplotlib.pyplot as plt
import numpy as np

# Se crea la primera grafica (grafica lineal)
plt.subplot(1, 4, 1) # se crea un subplot de 1 fila, 3 columnas y se selecciona la primera posicion
values = np.array([1,2,3,4,5]) # valores para la grafica lineal
plt.plot(values, label="Lineal", color='b') # grafica lineal con los valores dados y etiqueta Lineal y color azul

# Se crea la segunda grafica (grafica de barras)
plt.subplot(1, 4, 2) # se crea un subplot de 1 fila, 3 columnas y se selecciona la segunda posicion
values_x = ['a', 'b', 'c', 'd', 'e'] # valores para las barras en x
values_y = np.random.randint(low=1, high=10, size=5) # valores aleatorios para las barras en y
plt.bar(values_x, values_y, label="values", color="red") # grafica de barras con los valores dados en x y y y etiqueta values y color rojo 

# Se crea la tercera grafica (grafico de pastel)
plt.subplot(1, 4, 3) # se crea un subplot de 1 fila, 3 columnas y se selecciona la tercera posicion
values = np.random.randint(low=1, high=10, size=5) # valores aleatorios para el grafico de pastel
labels = ['A', 'B', 'C', 'D', 'E'] # etiquetas para las secciones del grafico de pastel
plt.pie(values, labels=labels, autopct='%1.1f%%') # grafico de pastel con los valores y etiquetas dadas y porcentajes de 1 decimal

#Se Crea La 4a grafica (histograma 2 dimensiones)
plt.subplot(1, 4, 4) #Se selecciona el subplot en la posicion 4
values_x = np.random.randint(low=1, high=100, size=10)# Valores aleatorios para barras en x
values_y = np.random.randint(low=1, high=100, size=10)#Valores aleaorios para barras en y
plt.hist2d(values_x, values_y, bins=30)

plt.show() # se muestra la imagen con las 4 graficas
```
